## Correlation with Heatmap

A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in Feature Selection

- It helps us to find the inter-relation between different features.
- For finding the inter-relation, our values needs to be integer or floating digits while for categorical features we will not able to find the correlation because of these are object type.
- The correlation coefficient ranges from –1 to 1 and that's also called the pearson's coefficient correlation.
- When it is close to 1, it means that there is a strong positive correlation; for example, the median_house_value (1.0) tends to go up when the median_income (0.69) goes up.
- When the coefficient is close to –1, it means that there is a strong negative correlation; you can see a small negative correlation between the latitude (-0.14) and the median_house_value (1.0) (i.e., prices have a slight tendency to go down when you go north).
- Finally, coefficients close to 0 mean that there is no linear correlation.

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
```

## JoinPlot
A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

## Univariate Analysis

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')
```

## Pair plot
A “pairs plot” is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
```

## Dist plot
Dist plot helps us to check the distribution of the columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde=False, bins=10)
```
